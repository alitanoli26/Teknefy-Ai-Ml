```python
# A set stores unique values (no duplicates allowed).
s1 = {1, 2, 3}
s2 = {3, 4, 5}

print(s1.union(s2))  # Combines all values from both sets (removes duplicates automatically).

print(s1.intersection(s2)) # Shows values that appear in BOTH sets.

# Shows values from s1 that are NOT in s2.
print(s1 - s2)
```
