```python
# Writing to a file:
# "w" mode means write mode. It creates a new file or overwrites if it exists.
with open("example.txt", "w") as file:
    file.write("Hello, My Name is Ali Hassan.I am Doing Bs In software Engineering ")  # Writing text inside the file.

# Reading from a file:
# "r" mode means read mode.
with open("example.txt", "r") as file:
    content = file.read()  # Reads the entire content of the file
    print(content)
```
