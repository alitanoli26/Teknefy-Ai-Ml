```python
class BankAccount:
    def __init__(self, balance):

        # This protects the data from misuse.
        self.__balance = balance

    def deposit(self, amount):
        # A PUBLIC method that safely changes the private data.
        self.__balance += amount

    def get_balance(self):
        # A PUBLIC method used to access hidden data safely.
        return self.__balance


# Creating object
acc = BankAccount(1000)
acc.deposit(500)    # modifying private data through method
print(acc.get_balance())  # reading private data through getter
```
