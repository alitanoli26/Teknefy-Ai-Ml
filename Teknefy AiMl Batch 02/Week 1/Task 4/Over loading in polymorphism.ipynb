```python
class Math:
    # Same method name but different number of parameters using default value.
    # Python doesn't support true overloading, so we fake it using =0.
    def add(self, a, b, c=0):
        return a + b + c

m = Math()
print(m.add(2, 3))     # uses a+b
print(m.add(2, 3, 4))  # uses a+b+c
```
