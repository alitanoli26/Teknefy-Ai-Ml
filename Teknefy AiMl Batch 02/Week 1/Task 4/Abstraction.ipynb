```python
from abc import ABC, abstractmethod


# It forces child classes to implement certain methods.
class Shape(ABC):

    # abstractmethod means this method MUST be written in all child classes.
    @abstractmethod
    def area(self):
        pass  # no body → because we hide the details

class Circle(Shape):
    def __init__(self, r):
        self.r = r  # radius

    # Implementing the abstract method.
    def area(self):
        # User doesn't need to know how area is calculated → abstraction
        return 3.14 * self.r * self.r


# Creating object of child class
c = Circle(5)
print(c.area())
```
