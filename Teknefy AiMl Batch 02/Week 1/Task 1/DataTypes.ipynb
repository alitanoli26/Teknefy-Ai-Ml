```python

# Integer:
# This is a whole number without decimals.
x = 26
print(type(x)) 
# type() is used to check the data type .


# Float:
# A number that contains a decimal point.
y = 2.3
print(type(y))


# String:
# A string is a sequence of characters (text).
s = "Ali"
print( type(s) )


# Boolean:
# Holds only two possible values: True or False.
flag = True
print(type(flag))



```
