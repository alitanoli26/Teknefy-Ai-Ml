```python
# A string is text stored inside quotes.
text = "My Name is Ali Hassan i am doing Python in jupyter notebook"


print(text.upper())  # Converts every letter in the string into UPPERCASE.


print(text.lower())  # Converts every letter in the string into lowercase.


print(text.title())  # Makes the first letter of each word capital (like a heading).


print(text.replace("Python", "Programming"))  # Replaces a specific word or letter with another.


print(len(text)) # Tells how many characters the string contains including spaces.


print(text[0])  # Strings work like arrays; [0] gives the first character.


print(text[-1])  # Negative index starts from the end; [-1] gives last character.
```
