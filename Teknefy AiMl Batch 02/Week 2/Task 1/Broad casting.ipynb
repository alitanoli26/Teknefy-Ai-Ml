```python
import numpy as np
#by broad casting we can add or multiply the matrices of different shapes
arr=np.array([1,2,3,4])
arr1=np.array([[10],[20],[30]])
print(f"Addition of Matrices \n{arr+arr1}") #Adding two metrices of different shape
```

```python
arr=np.array([1,2,3,4])
arr1=np.array([[10],[20],[30]])
print(f"Multiplication of Matrices \n {arr*arr1}")  #Multiplying two matrices of different shape
```
