```python
import pandas as pd
# Load the CSV file
df = pd.read_csv("data.csv")
# Using apply
df["Grade"] = df["Marks"].apply(lambda x: "A" if x > 85 else "B")
print(df.head())
```

```python
# Using map
dept_map = {"CS": "Computer Science", "IT": "Information Tech", "SE": "Software Eng", "AI": "Artificial Intelligence"}
df["Dept_full"] = df["Department"].map(dept_map)
print(df.head())
```

```python
# Using transform
df["Double_Age"] = df["Age"].transform(lambda x: x * 2)

print(df.head())
```
